```python
#grouping the body style using the average price of each, since we have just 5 body styles,using the groupby function
df_group1= df_group1.groupby('body-style',as_index=False).mean()
```

```python
df_group2=df[['body-style','price']]
df_group2
```

```python
#grouping by the sum
df_group2= df_group2.groupby('body-style',as_index=False).sum()
df_group2
```

```python
plt.bar(df_group2['body-style'],df_group2['price'],color='darkblue')
plt.title('BODY STYLE Price')
plt.ylabel('Sum of Prices')
plt.xlabel('Body Style',color='darkred')
plt.show()
#le6---exponential raise tpower of 6
```
